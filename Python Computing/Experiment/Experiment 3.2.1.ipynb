```python
def Ackermann(m, n):
    if not m:
        return n + 1
    if m > 0 and not n:
        return Ackermann(m - 1, 1)
    if m > 0 and n > 0:
        return Ackermann(m - 1, Ackermann(m, n - 1))


print("Ackermann(3, 4) = %d." % Ackermann(3, 4))
```

```python
def DeleteRepeat(elems):
    if str(type(elems)) == "<class 'list'>":
        result = []
        noRepeat = set()
        for elem in elems:
            originalLength = len(noRepeat)
            noRepeat.add(elem)
            if len(noRepeat) != originalLength:
                result.append(elem)
        return result
    else:
        result = {}
        noRepeat = set()
        for elem in elems:
            originalLength = len(noRepeat)
            noRepeat.add(elems[elem])
            if len(noRepeat) != originalLength:
                result[elem] = elems[elem]
        return result


repeatList = [0, 0, 1, 2, 2, 3, 4, 4, 4, 4, 5, 6, 7, 8, 9, 9, 9]
print("The original list is: %s" % repeatList)
print("After deleted repeat, it is: %s" % DeleteRepeat(repeatList))
repeatDictionary = {"a": 1, "b": 2, "c": 3, "d": 4, "e": 1}
print("The original dictation is: %s" % repeatDictionary)
print("After deleted repeat, it is: %s" % DeleteRepeat(repeatDictionary))
```
